```python
import numpy as np 
from scipy.fftpack import dct, idct
import math
```

```python
def compute_xk(k):
    res = []
    for i in range(k):
        x_k = ( (i-8)**3 ) / 8
        res.append(x_k)
    return res
```

```python
def compute_yk(k):
    res = []
    for i in range(k):
        y_k = math.cos( (3*i+1)*math.pi/16 + 1 ) + math.sin( (5*i+1) * math.pi/16 )
        res.append(y_k)
    return res
```

```python
n = 16
x = compute_xk(n)
y = compute_yk(n)
```

```python
X = np.array(dct(x))
Y = np.array(dct(y))
```

```python
X_hat = X.copy()
X_hat[-10:] = 0

Y_hat = Y.copy()
Y_hat[-10:] = 0
```

```python
x_hat = idct(X_hat)
y_hat = idct(Y_hat)
```

```python
X_table = [X, X_hat]
x_table = [x, x_hat]
Y_table = [Y, Y_hat]
y_table = [y, y_hat]
```
